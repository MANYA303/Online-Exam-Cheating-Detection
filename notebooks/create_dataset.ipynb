```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)

students = 200

data = pd.DataFrame({
    "student_id": [f"S{1000+i}" for i in range(students)],
    "exam_duration": np.random.normal(120, 10, students).astype(int),
    "tab_switch_count": np.random.poisson(2, students),
    "mouse_movements": np.random.normal(300, 50, students).astype(int),
    "keypress_count": np.random.normal(500, 100, students).astype(int),
    "idle_time": np.random.normal(60, 20, students).astype(int),
    "ip_change": np.random.choice([0, 1], students, p=[0.9, 0.1]),
    "copy_paste_count": np.random.poisson(0.5, students)
})

# Inject cheating behavior
for i in random.sample(range(students), 10):
    data.loc[i, ["tab_switch_count","idle_time","copy_paste_count"]] = [25, 300, 8]

data.to_csv("../data/exam_logs.csv", index=False)
data.head()
```
